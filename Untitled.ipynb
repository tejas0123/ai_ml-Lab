```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
print(mean_x)
print(mean_y)
n = len(X)
print(n)
```

```python
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x)*(Y[i] - mean_y)
    denom += (X[i]-mean_x)**2
b1 = numer/denom
b0 = mean_y - (b1*mean_x)
print("Coefficients")
print(b1, b0)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1*x
```

```python
plt.plot(x, y, color = '#58b970', label = 'Regression line')

#plotting scatter plots
plt.scatter(X, Y, c = '#ef5423', label = 'scatter plot')
plt.xlabel('Head size in cm^3')
plt.ylabel('Brain weight in grams')
plt.legend()
plt.show()
```

```python
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = b0 + (b1*X[i])
    ss_tot += (Y[i]-mean_y)**2
    ss_res += (Y[i]-y_pred)**2
r2 = 1-(ss_res/ss_tot)
print("The R^2 score is ")
print(r2)
```
