```python
#import datasets and libraries
import pandas as pd
df = pd.read_csv('ch1ex1.csv')
points = df.values
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3)
model.fit(points)
labels = model.predict(points)
import matplotlib.pyplot as plt
```

```python
#Assign column 0 of points to xs
xs = points[:,0]
ys = points[:,1]
```

```python
#Make a scatter plot of xs and ys, specifying the c=labels keyword arguments to color the points by their cluster label. 
plt.scatter(xs, ys, c = labels)
plt.show()
```

```python
centroids = model.cluster_centers_
```

```python
#assign column 0 of centroids to centroids _x and column 1 of centroids to centroids_y
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]
```

```python
plt.scatter(xs, ys, c = labels)
plt.scatter(centroids_x, centroids_y, marker = "X", s = 200)
plt.show()
```
